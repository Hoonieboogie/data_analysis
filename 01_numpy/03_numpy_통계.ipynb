# Numpy 통계 관련

```python
import numpy as np
```

## 중심 경향 측도 (Central Tendancy)

```python
height = np.array([160, 170, 180, 172, 164, 152, 148, 192, 187, 177, 166, 152, 156, 176, 189, 160]) 
```

```python
# 평균값 mean
print(np.mean(height))
```

```python
# 중위값 median
print(np.sort(height))
print(np.median(height))
```

```python
# 최빈값 mode (numpy에는 바로 구하는 함수가 없음)
values, count = np.unique(height, return_counts=True) # 중복없이 array & 각각의 count의 담은 arraay (return_counts= True) 
max_count = np.max(count)

print(values)
print(count)
print(count == max_count)
print(values[count == max_count])
```

---

## 산포(Dispersion) 측도


```python
# 분산 (variance) : {(값-평균)^2} / N
print(np.mean(np.square(height - np.mean(height))))
print(np.var(height))
```

```python
# 표춘 편차 (standard deviation)
print(np.sqrt(np.var(height)))
print(np.std(height))
```

```python
# 범위 (range)
print(np.max(height) - np.min(height))
```

# 확률 분포

```python
#!pip install matplotlib
import matplotlib.pyplot as plt
```

```python
# 표준 정규분포 standard normal dist.: mean(평균) 0, sd(표준편차):1

# 만드는 방법 1: np.random.randn(shape) => ㄴnormal distribution based random num generator (무조건 m:0, sd:1)
# args 없으면 숫자 하나
arr = np.random.randn(1000)
plt.hist(arr, bins=1000)
plt.show()
```

```python

# 정규분포: mean(평균) 0, sd(표준편차):1

# 만드는 방법 2: np.random.normal(loc, scale, size) =>  normal distribution based random num generatorarr 
                                # loc: mean
                                # scale: sd
                                # size: 개수 또는 shape
arr = np.random.normal(10,1,(10,10))
print(arr.shape)

plt.hist(arr, bins=100)
plt.show()
```

```python
# 균등 분포 (uniform distribution)
# np.random.rand(num) => np.random.radn()이랑 달라요
arr = np.random.rand(1000)
arr = np.random.rand(2,3)
print(arr)


plt.hist(arr, bins=10)
plt.show
```

```python
# 균등 분포 (uniform distribution)
# np.random.randint(start,end,size)

np.random.seed(0) # 난수 계산 시작값 seed를 0으로 고정

arr = np.random.randint(1, 100, 100)
arr = np.random.randint(1, 100, (2,3))
print(arr)
print(np.max(arr), np.min(arr))

plt.hist(arr, bins=100)
plt.show()
```
