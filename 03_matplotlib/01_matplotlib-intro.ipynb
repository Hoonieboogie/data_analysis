# Matplotlib
기본적으로 line graph를 그리게 해주는 패키지

```python
#%pip install matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 힌글 폰트 사용을 위한 설정
import matplotlib.font_manager as fm
import matplotlib

font_path = 'C:/Windows/Fonts/gulim.ttc'
font = fm.FontProperties(fname=font_path).get_name()
matplotlib.rc('font', family=font) # 폰트 전역 설정
```

## 기본 사용법

```python
x = [1,2,3,4,5,6]
y = [7,8,9,10,11,12]

plt.plot(x,y)
plt.title("Test Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

### Figure | Axes
- Figure: 전체 그림이 그려지는 컨테이너
- Axes: 그래프가 실제로 그려지는 영역 (좌표, 데이터)

```python
fig = plt.figure(figsize=(6,4), facecolor='green') # fig 객체 생성
print(fig) 

ax = plt.gca() # gca(get current axis) -> 실제 그래프가 그려지는 영역
ax.plot(x,y)
print(ax)

plt.title("Test Graph (Colored)")
plt.xlabel("X Value")
plt.ylabel("Y Value")

plt.show()
```

```python
#plt.subplots(2,2) # 행 2 열 3

# subplots()는 fig와 ax 객체 2개 반환
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(10,6)) # 이렇게도 가능 (행 2 열 3)

print(fig)
print(ax)

plt.show()
```

### plot 메서드 ➡️ x,y 인자를 가짐

```python
# 1. 파이썬의 List 가능
```

```python
# 2. ndarray 가능
x = np.arange(1,11)
y = x*2
x,y
```

```python
# 3. Series 가능
x = pd.Series(x)
y = pd.Series(y)
x,y
```

```python
# 4. DataFrame 가능
x = pd.DataFrame({'x':x})
y = pd.DataFrame({'y':y})
x,y
```

```python
plt.plot(x,y)
plt.show()
```

### plot 메서드의 파라미터

```python
plt.figure(figsize=(6,4))

plt.plot(
    x,y,
    color='red', # line color
    marker = 'o', # o, x, ^, ., *, H, P, D,  ...  => 하나 plot의 여러개 선 넣어야 하면 유용
    markersize = 5,
    linestyle = '--',  # dotted, dashdot, dashed, None, ...
    linewidth = 20
)

plt.show()
```

### plt 속성 설정

```python
# label: 기본적으로 한글 '폰트' 지원하지 않음
plt.plot(x,y, marker='o')
plt.title('졸리다 집가자')
plt.xlabel('x축이다 ㅅㄱ')
plt.ylabel('y축임ㅊㅊ')
plt.show()
```

```python
# ticks (눈금)
plt.figure(figsize=(3,3))

x = np.arange(1,101)
y = np.linspace(2,200, 100) # 2부터 200까지 100개의 요소

plt.xticks(ticks=np.arange(0,101, 10), rotation=90) # 0부터 100까지 10개 간격
plt.yticks(rotation=45)

plt.plot(x,y, linestyle='--')
plt.show()
```

```python
# lim: 축 보여지는 범위 재지정
plt.figure(figsize=(3,2))

x = np.arange(1,101)
y = np.linspace(2,200, 100) # 2부터 200까지 100개의 요소

plt.xlim(1,200)
plt.ylim(1,1000)

plt.plot(x,y, linestyle='--')
plt.show()
```

```python
# legend (범례)
plt.figure(figsize=(3,3))

x = np.arange(1,101)
y = np.linspace(2,200,100)
y2 = np.linspace(3,300,100)

plt.plot(x,y, label='파랭이')
plt.plot(x,y2, label='감귤이')

# 범례 위치 loc
# - 기본값: 'best'
# - center
# (upper | lower) (left | right)
plt.legend(loc='upper left') # loc 기본값은 'best'

plt.show()
```

## 그래프 여러 개 그리기

```python
# 여러 종류 그래프 동시에 그리기
plt.figure(figsize=(3,3))

x = np.arange(1,11)
y = x*3
y2 = x*2

plt.plot(x,y, label='빨간선', color = 'red', marker='o')
plt.bar(x,y2, label='파란바', color='blue')

plt.legend() 
plt.grid()

plt.show()
```

```python
# 여러 그래프 각각 그리기 -> subplots()
fig, ax = plt.subplots(1,2, figsize=(10,5)) # 한 row에 2개 그래프 그리기

x11 = np.arange(1,11)
y11 = x11*2
y12 = x11**2


x21 = np.arange(11,20)
y21 = x21*2
y22 = (x21*2) * 0.6

ax[0].plot(x11,y11, color='red', label='빨간선')
ax[0].plot(x11,y12, color='green', label='초록선')
# 그래프 설정 (왼쪽)
ax[0].set_title("왼쪽")
ax[0].set_xlabel("X1")
ax[0].set_ylabel("Y1")
ax[0].legend()
ax[0].grid()

ax[1].bar(x21,y21, label='파란바')
ax[1].bar(x21,y22, label='주황바')
# 그래프 설정 (오른쪽)
ax[1].set_title('오른쪽')
ax[1].set_xlabel("X2")
ax[1].set_ylabel("Y2")
ax[1].legend()
ax[1].grid()





ax[1].legend()

plt.show()
```
